```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Cargar el archivo CSV
archivo_csv = "BD.csv"  # Cambia esto a la ruta de tu archivo real
df = pd.read_csv(archivo_csv)

# Convertir la columna "Usage" a flotantes (removiendo el smbolo '%')
df["Usage"] = df["Usage"].str.rstrip('%').astype(float)

# Crear un ndice ficticio como variable independiente
df = df.sort_values(by="Usage", ascending=False).reset_index(drop=True)
df["Ranking"] = np.arange(1, len(df) + 1)

# Separar las variables independiente (Ranking) y dependiente (Usage)
X = df[["Ranking"]].values  # Variable independiente
y = df["Usage"].values       # Variable dependiente

# Crear y ajustar el modelo de regresin lineal
model = LinearRegression()
model.fit(X, y)

# Generar predicciones
y_pred = model.predict(X)

# Mostrar los coeficientes del modelo
print("Pendiente (coeficiente):", model.coef_[0])
print("Intercepto:", model.intercept_)

# Graficar los datos reales y la regresin lineal
plt.figure(figsize=(12, 6))
plt.scatter(df["Ranking"], df["Usage"], color="blue", label="Datos reales")
plt.plot(df["Ranking"], y_pred, color="red", label="Regresin Lineal")

# Etiquetas y ttulo
plt.xlabel("Ranking de Bases de Datos")
plt.ylabel("Porcentaje de Uso (%)")
plt.title("Regresin Lineal sobre Popularidad de Bases de Datos")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
