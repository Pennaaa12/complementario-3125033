```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Ruta de tu archivo CSV
archivo_csv = "Lenguajesdeprogramacion.csv"  # Cambia esto a la ruta de tu archivo real

# Cargar el archivo CSV
df = pd.read_csv(archivo_csv)

# Asegurarse de que la columna "Percentage" est en formato numrico
df["Percentage"] = df["Percentage"].astype(float)

# Ordenar los lenguajes por porcentaje de uso de mayor a menor
df = df.sort_values(by="Percentage", ascending=False).reset_index(drop=True)

# Crear un ndice ficticio para representar el ranking
df["Ranking"] = np.arange(1, len(df) + 1)

# Separar las variables independiente (Ranking) y dependiente (Percentage)
X = df[["Ranking"]].values  # Variable independiente
y = df["Percentage"].values  # Variable dependiente

# Crear y ajustar el modelo de regresin lineal
model = LinearRegression()
model.fit(X, y)

# Generar predicciones
y_pred = model.predict(X)

# Mostrar los coeficientes del modelo
print("Pendiente (coeficiente):", model.coef_[0])
print("Intercepto:", model.intercept_)

# Graficar los datos reales y la regresin lineal
plt.figure(figsize=(12, 6))
plt.scatter(df["Ranking"], df["Percentage"], color="blue", label="Datos reales")
plt.plot(df["Ranking"], y_pred, color="red", label="Regresin Lineal")

# Etiquetas y ttulo
plt.xlabel("Ranking de Lenguajes de Programacin")
plt.ylabel("Porcentaje de Uso (%)")
plt.title("Regresin Lineal sobre Popularidad de Lenguajes de Programacin")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
